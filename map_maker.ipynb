```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
env = np.zeros((100, 100, 100))
env[5:10, 5:10, 0:100] = 1
# env[9:12, 10:30, 10:90] = 1
# env[45:55, 45:55, 0:100] = 1
env[60:95, 80:85, 0:100] = 1
# env[20:25, 5:100, 0:80] = 1
env[40:80, 35:40, 0:40] = 1
env[40:80, 35:40, 60:100] = 1
env[35:40, 35:40, 0:100] = 1
env[80:85, 35:40, 0:100] = 1
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x, y, z = env.nonzero()
ax.scatter(x, y, z, zdir='z', color='b')

ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_zlim(0, 100)

ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_zlabel('Y')
```
